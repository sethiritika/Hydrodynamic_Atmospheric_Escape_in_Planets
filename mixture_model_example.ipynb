Fill in planet/star parameters below in cgs units. By default the model returns mass-loss rate normalized to the planet's surface density, $\dot M/\rho_s$ in $\rm cm^3$ $\rm s^{-1}$. 

If you supply `rho_surface` in g $\rm cm^{-3}$, the notebook also returns absolute $\dot M$ in g $\rm s^{-1}$.

If rho_surface is unknown but you want to assume a pressure level, the notebook estimates rho_surface from the ideal gas law. 

With our definition:
     
$\rm cs^2$ = $\gamma$ * $k_B$ * T / ($\mu$ * $m_p$).   #cs is the speed of sound

the ideal gas law gives:

rho_surface = $\gamma$ * P_surface / $\rm cs^2$

For the isothermal gas $\gamma = 1$: 

P_surface = rho_surface * $\rm cs^2$
