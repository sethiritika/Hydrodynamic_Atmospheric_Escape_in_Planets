# RLOF Mixture Model Mass-Loss Calculator

Fill in planet/star parameters below in cgs units. By default the model returns $\dot M/\rho_s$ in cm$^3$ s$^{-1}$. If you supply `rho_surface` in g cm$^{-3}$, the notebook also returns absolute $\dot M$ in g s$^{-1}$.

```python
import numpy as np
from mixture_model import mass_loss_rates, rho_from_pressure

M_sun = 1.98847e33
M_earth = 5.9722e27
R_earth = 6.371e8
AU = 1.495978707e13
```

## User Inputs

Set `gamma=1` for the isothermal sound speed used by the model. Change `rho_surface` if you have an atmospheric density estimate at the effective surface.

```python
Mstar = 1.0 * M_sun
Mp = 10.0 * M_earth
Rp = 3.0 * R_earth
a = 0.05 * AU
T = 1.0e4      # K
mu = 0.6       # dimensionless, in proton masses
gamma = 1.0    # isothermal

# Option 1: leave as None to get Mdot/rho_s in cm^3/s
rho_surface = None

# Option 2: estimate rho_s from an assumed pressure level
# P_surface = 1.0  # dyn/cm^2 = 1 microbar
# rho_surface = rho_from_pressure(P_surface, T, mu, gamma)
```

```python
rates = mass_loss_rates(Mstar, Mp, Rp, a, T, mu, gamma, rho_surface=rho_surface)

unit = "g/s" if rho_surface is not None else "cm^3/s (per unit rho_s)"
for name, value in rates.items():
    print(f"{name:22s}: {value:.4e} {unit}")
```
